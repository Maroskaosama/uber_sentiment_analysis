```python
from google_play_scraper import reviews
import pandas as pd
import time

all_reviews = []
count = 0

while True:
    result, continuation_token = reviews(
        'com.ubercab',  # Uber app ID
        lang='ar',      # Arabic language
        country='eg',   # Egypt
        count=200,
        filter_score_with=None,
        continuation_token=None if count == 0 else continuation_token
    )
    all_reviews.extend(result)
    print(f"Scraped {len(all_reviews)} reviews...")
    count += 1
    time.sleep(1)
    if continuation_token is None or count >= 10:
        break

df = pd.DataFrame(all_reviews)
df = df[['userName', 'content', 'score', 'at']]
df.dropna(subset=['content'], inplace=True)
df.to_csv("uber_reviews_ar.csv", index=False, encoding="utf-8-sig")
print("✅ Reviews saved to 'uber_reviews_ar.csv'")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_top_n_words(corpus, n=20):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)

    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

    return words_freq[:n]
```

```python
def plot_top_words(word_freq, title):
    words = [w for w, c in word_freq]
    counts = [c for w, c in word_freq]

    plt.figure(figsize=(10, 5))
    sns.barplot(x=counts, y=words, palette="viridis")
    plt.title(title)
    plt.xlabel("Frequency")
    plt.ylabel("Word")
    plt.show()
```

```python
# Positive words
top_pos = get_top_n_words(df[df['label'] == 'positive']['clean_text'])
plot_top_words(top_pos, "Top Words in Positive Reviews")

# Negative words
top_neg = get_top_n_words(df[df['label'] == 'negative']['clean_text'])
plot_top_words(top_neg, "Top Words in Negative Reviews")

# (Optional) Neutral words
top_neu = get_top_n_words(df[df['label'] == 'neutral']['clean_text'])
plot_top_words(top_neu, "Top Words in Neutral Reviews")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('/content/uber_reviews_ar.csv')

# Make sure the text cleaning function is defined (assuming it's in a previous cell)
import re
import pyarabic.araby as araby
from nltk.corpus import stopwords

def normalize_arabic(text):
    text = araby.strip_tashkeel(text)  # Remove diacritics
    text = re.sub(r'(.)\1+', r'\1\1', text)  # Reduce character elongation
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # Remove non-Arabic characters
    return text

arabic_stopwords = set(stopwords.words("arabic"))

# Apply text cleaning to the 'content' column to create 'clean_text'
df["clean_text"] = df["content"].apply(lambda x: " ".join(
    [word for word in normalize_arabic(x).split() if word not in arabic_stopwords]
))

# Create sentiment labels based on 'score' column (assuming score is numeric rating)
# Re-creating the 'label' column as in the original code before the error
def label_sentiment(score):
    if score >= 4:
        return 'positive'
    elif score <= 2:
        return 'negative'
    else:
        return 'neutral'

df['label'] = df['score'].apply(label_sentiment)
df.dropna(subset=['clean_text'], inplace=True) # Use clean_text for dropping rows
df.reset_index(drop=True, inplace=True)


# Split into train/test - Use 'label' as the target variable as defined before
X_train, X_test, y_train, y_test = train_test_split(df['clean_text'], df['label'], test_size=0.2, random_state=42)

# Vectorize text
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Train model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Plot confusion matrix
ConfusionMatrixDisplay.from_estimator(model, X_test_tfidf, y_test, cmap="Purples")
plt.title("Confusion Matrix")
plt.show()
```

```python
import re
import pyarabic.araby as araby  # if installed, else you can use basic regex

def preprocess_arabic(text):
    if not isinstance(text, str):
        return ""

    # Normalize Arabic letters (remove tatweel/kashida, normalize Alef variants)
    text = re.sub(r"[إأآا]", "ا", text)
    text = re.sub(r"ى", "ي", text)
    text = re.sub(r"ؤ", "ء", text)
    text = re.sub(r"ئ", "ء", text)
    text = re.sub(r"ة", "ه", text)
    text = re.sub(r"\ـ", "", text)  # Remove tatweel

    # Remove diacritics
    diacritics = re.compile("""
                             ّ    | # Shadda
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
    text = re.sub(diacritics, '', text)

    # Remove non-Arabic letters, digits, punctuations, emojis etc.
    text = re.sub(r"[^\u0600-\u06FF\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Usage example
df['clean_text'] = df['score'].apply(preprocess_arabic)
```

```python
# Predict probabilities instead of just labels
probs = model.predict_proba(X_test_tfidf)  # shape: (n_samples, n_classes)

# Get max probability and predicted class index for each sample
max_probs = probs.max(axis=1)
predicted_classes = model.classes_[probs.argmax(axis=1)]

# Set a confidence threshold, e.g. 0.7
threshold = 0.7

# Apply threshold: if confidence < threshold, label as "uncertain" or "neutral"
final_preds = []
for prob, pred in zip(max_probs, predicted_classes):
    if prob < threshold:
        final_preds.append("neutral")  # or "uncertain"
    else:
        final_preds.append(pred)

# Example: show results for first 10 samples
for i in range(10):
    print(f"Text: {X_test.iloc[i]}")
    print(f"Predicted: {final_preds[i]} with confidence {max_probs[i]:.2f}\n")
```
